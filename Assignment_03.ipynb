```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("jahias/microsoft-adventure-works-cycles-customer-data")

print("Path to dataset files:", path)
```

```python
import pandas as pd
df = pd.read_csv(path + "/AWCustomers.csv")

df.info()
display(df.head())
display(df.describe())
display(df.nunique())
```

```python
from datetime import datetime

# 1. Calculate the age of each customer
df['BirthDate'] = pd.to_datetime(df['BirthDate'])
today = datetime.now()
df['Age'] = (today - df['BirthDate']).dt.days // 365

# 2. Create a new categorical column 'AgeGroup'
bins = [0, 25, 40, 60, df['Age'].max()]
labels = ['0-25', '26-40', '41-60', '61+']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=True, include_lowest=True)

# 3. Analyze the distribution of customers by 'Gender', 'MaritalStatus', 'Education', and 'Occupation'.
print("Distribution by Gender:")
display(df['Gender'].value_counts())
print("\nDistribution by MaritalStatus:")
display(df['MaritalStatus'].value_counts())
print("\nDistribution by Education:")
display(df['Education'].value_counts())
print("\nDistribution by Occupation:")
display(df['Occupation'].value_counts())

# 4. Calculate the average 'YearlyIncome' for each 'Occupation' and 'Education' level.
print("\nAverage YearlyIncome by Occupation:")
display(df.groupby('Occupation')['YearlyIncome'].mean().sort_values(ascending=False))
print("\nAverage YearlyIncome by Education:")
display(df.groupby('Education')['YearlyIncome'].mean().sort_values(ascending=False))

# 5. Analyze the relationship between 'NumberCarsOwned', 'NumberChildrenAtHome', and 'TotalChildren' with 'YearlyIncome'.
print("\nAverage YearlyIncome by NumberCarsOwned:")
display(df.groupby('NumberCarsOwned')['YearlyIncome'].mean())
print("\nAverage YearlyIncome by NumberChildrenAtHome:")
display(df.groupby('NumberChildrenAtHome')['YearlyIncome'].mean())
print("\nAverage YearlyIncome by TotalChildren:")
display(df.groupby('TotalChildren')['YearlyIncome'].mean())

# 6. Explore the geographical distribution of customers by 'CountryRegionName', 'StateProvinceName', and 'City'.
print("\nDistribution by CountryRegionName:")
display(df['CountryRegionName'].value_counts())
print("\nDistribution by StateProvinceName (Top 10):")
display(df['StateProvinceName'].value_counts().head(10))
print("\nDistribution by City (Top 10):")
display(df['City'].value_counts().head(10))
```

 Add a new column that categorizes employees based on their performance rating:

i. Excellent for ratings >= 4.5

ii. Good for ratings >= 4.0 but < 4.5

iii. Average for ratings < 4.0  

```python
def categorizes_employees(ratings):
  if ratings >= 4.5:
   return 'excellent'
  if ratings< 4.5:
   return "good"
  else:
   return "average"

emp['Performance'] = emp['Rating'].apply(categorizes_employees)
emp
```
